```python
import numpy as np
import matplotlib.pyplot as plt

true_mean = 5
samples = np.random.normal(loc=true_mean, scale=2, size=10000)
empirical_means = np.array([np.mean(samples[:n]) for n in range(1, len(samples))])

distances=np.abs(empirical_means-true_mean)

plt.plot(empirical_means, label='Empirical mean')
plt.axhline(true_mean, color='r', linestyle='--', label='True mean')
plt.xlabel('Number of samples')
plt.ylabel('Empirical mean')
plt.title('Law of Large Numbers')
plt.legend()
plt.show()

plt.scatter(range(1,distances.shape[0]+1),distances,s=0.5)
plt.scatter(range(1,distances.shape[0]+1),1/np.sqrt(range(1,distances.shape[0]+1)),s=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n_values = np.arange(10, 10000, 100)
num_reps = 1000
std_dev = 2
true_mean = 5

errors = []
for n in n_values:
    estimates = [np.mean(np.random.normal(loc=true_mean, scale=std_dev, size=n)) for _ in range(num_reps)]
    abs_error = np.mean(np.abs(np.array(estimates) - true_mean))
    errors.append(abs_error)

plt.plot(n_values, errors, label='Avg absolute error over reps')
plt.plot(n_values, std_dev/np.sqrt(n_values)*np.sqrt(2/np.pi), label=r'$1/\sqrt{n}$ scaled')
plt.legend()
plt.title("Empirical convergence to 1/sqrt(n)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

true_mean = 5
samples = np.random.exponential(true_mean, size=10000)
empirical_means = np.array([np.mean(samples[:n]) for n in range(1, len(samples))])
distances=np.abs(empirical_means-true_mean)
plt.plot(empirical_means, label='Empirical mean')
plt.axhline(true_mean, color='r', linestyle='--', label='True mean')
plt.xlabel('Number of samples')
plt.ylabel('Empirical mean')
plt.title('Law of Large Numbers')
plt.legend()
plt.show()

plt.scatter(range(1,distances.shape[0]+1),distances)
plt.scatter(range(1,distances.shape[0]+1),1/np.sqrt(range(1,distances.shape[0]+1)))
plt.show()
```

```python
import numpy as np

states = ["S0: Confusion", "S1: Trying", "S2: Breakthrough", "S3: Plku!"]
P = np.array([
    [0.5, 0.4, 0.1, 0.0],
    [0.2, 0.5, 0.2, 0.1],
    [0.0, 0.2, 0.5, 0.3],
    [0.0, 0.0, 0.0, 1.0]
])

np.random.seed(42)
current = 0
trajectory = [states[current]]

for _ in range(20):  # simulate 20 steps
    current = np.random.choice([0,1,2,3], p=P[current])
    trajectory.append(states[current])
    if current == 3:
        break  # once Plku appears, you're done 

print("Your motivational journey:")
for step, state in enumerate(trajectory):
    print(f"Step {step}: {state}")
```
