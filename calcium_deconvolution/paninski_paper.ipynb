```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_spike_train(rate, duration):
    return np.random.poisson(rate, duration)

print(generate_spike_train(10, 100))
```

```python
s_t=generate_spike_train(10, 100) #denote by st the number of spikes that the neuron fired at the t-th timestep, t = 1, â€¦, T

def autoregressive_calcium_trace(spike_train, gamma=np.array([0.01,0.2])):
    c_t=[0,0]
    for t in range(len(spike_train)):
        c=np.dot(np.array(c_t[-2:]),gamma)+spike_train[t]
        c_t.append(c)
    return np.array(c_t)

print(autoregressive_calcium_trace(s_t))
```

```python
trace=autoregressive_calcium_trace(s_t)

plt.plot(trace)
```
