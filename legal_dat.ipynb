```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load data
dat = pd.read_csv('/home/maria/Documents/LegalData/readability_scores.csv')

# Set figure size
plt.figure(figsize=(10, 6))

# Scatter plot with small points for visibility
plt.scatter(dat['year_filed'], dat['FleschReadingEaseScore'], alpha=0.5, s=0.5, c='orange', label='Data Points')

# Regression plot with 95% confidence interval
sns.regplot(x='year_filed', y='FleschReadingEaseScore', data=dat, scatter=False, ci=95, color='blue', line_kws={'label':"Regression Line"})

# Labels and title
plt.xlabel('Year Filed')
plt.ylabel('Flesch-Kincaid Reading Ease Score')
plt.title('US Supreme Court Opinions Flesch-Kincaid Reading Ease Score by Year Filed')

# Show legend
plt.legend()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/home/maria/Documents/LegalData/readability_scores.csv"  # Update if needed
dat = pd.read_csv(file_path)

# Ensure year is numeric
dat["year_filed"] = pd.to_numeric(dat["year_filed"], errors="coerce")

# Group by year and compute the mean Flesch-Kincaid score
yearly_avg = dat.groupby("year_filed")["FleschReadingEaseScore"].mean().reset_index()

# Set figure size
plt.figure(figsize=(10, 6))

# Line plot of the yearly average scores
sns.lineplot(x="year_filed", y="FleschReadingEaseScore", data=yearly_avg, marker="o", color="blue")

# Labels and title
plt.xlabel("Year Filed")
plt.ylabel("Average Flesch-Kincaid Reading Ease Score")
plt.title("US Supreme Court Opinions: Average Flesch-Kincaid Score by Year")

# Show grid for better readability
plt.grid(True)

# Show plot
plt.show()
```
