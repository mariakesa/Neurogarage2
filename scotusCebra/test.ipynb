```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dat=np.load('embs.npy')
```

```python
plt.plot(dat[:100,1])
```

```python
from sklearn.manifold import TSNE   

X_embedded = TSNE(n_components=2).fit_transform(dat[:1000,:])
```

```python
plt.scatter(X_embedded[:,0],X_embedded[:,1])
plt.show()
```

```python
import umap

embedding = umap.UMAP().fit_transform(dat[:1000,:])
```

```python
plt.scatter(embedding[:,0],embedding[:,1])
```
