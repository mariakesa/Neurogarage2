```python
import math

f=lambda x: math.sqrt(3-5*x+x**2+x**3)/(x-1)

x=[1.1,1.01, 1.001, 1.0001]

for i in x:
    print(f(i))
```

```python
import matplotlib.pyplot as plt
import numpy as np
ys=[]
xs=np.arange(-2,2,0.001)
xs

for i in xs:
    ys.append(f(i))

plt.plot(xs,ys)
```
